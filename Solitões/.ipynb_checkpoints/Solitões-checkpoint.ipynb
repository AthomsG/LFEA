Definem-se as funções auxiliares

```python
def read_file(file_path, columns = 5):
    with open(file_path, 'r') as file:
        dados = file.readlines()

    cor1_x  = list()
    cor1_y  = list()

    cor2_x  = list()
    cor2_y  = list()

    tempo   = list()

    for linha in dados[2:]:
        var      = linha.split()
        
        if len(var) == columns:
            tempo   += [float(var[0].replace(',','.'))]

            cor1_x  += [float(var[1].replace(',','.'))]
            cor1_y  += [float(var[2].replace(',','.').replace('\x00', ''))]
            
            if columns > 3:
                cor2_x  += [float(var[3].replace(',','.'))]
                cor2_y  += [float(var[4].replace(',','.').replace('\x00', ''))]
    if columns > 3:
        return tempo, cor1_x, cor1_y, cor2_x, cor2_y
    return tempo, cor1_x, cor1_y
```

```python
def chi_sq(obs, exp, err):
    soma = 0
    assert(len(obs) == len(exp)) #ERRO DE DIMENSÕES
    #raise TypeError("Erro de dimensões")
    for i in range(len(obs)):
        soma += pow((obs[i] - exp[i])/err[i], 2)
    return soma
        
```

# CORREÇÃO DOS DADOS PROVENIENTES DO TRACKER

Dedução da expressão no Logbook

```python
def correct(cor1_y, cor1_x, h): 
    eta = []
    beta = []

    theta = 0.927
    h_0=6
    r_z_0 = 0.021

    for i in range(len(cor1_x)):
        beta.append(np.arctan(cor1_x[i]/(cor1_y[i]+2*(r_z_0+(h-h_0)))))
        if (np.sin(2*beta[i]) == 0):
            eta.append((cor1_y[i])/2)
        else:
            eta.append(cor1_x[i]/np.sin(2*beta[i])-r_z_0-(h-h_0))
    return eta
```

## 1 Solitão

```python
def fit_func(t, a, x, c):
    h = 0.06
    L = (np.sqrt(4*pow(h, 3)/(3*a)))
    return a*pow(1/np.cosh(x/L - c*t/L), 2)

def fit_peak(t, a, b, c):
    return 2*a*c*np.tanh(b - c*t)*fit_func(t, a, b, c)
```

```python
x = np.linspace(-4, 4, 1000)
plt.plot(tempo, ccor1_y)

#plt.axis('off') 
plt.plot(x, fit_func(x, 0.01/1.5, 1, 0.6), label='fit function')
#plt.plot(x, fit_peak(x, 1, -2, 1), label='derivative')
plt.legend(bbox_to_anchor=(1, 1), loc=1, frameon=False, fontsize=17)
```

# Modelo para 2 solitões

\begin{equation}
    \begin{array}{c}
u (x, t)=2\left(c_{1}-c_{2}\right)\left(c_{1} \cosh ^{2}\left(\sqrt{c}_{2} \xi_{2} / 2\right)+c_{2} \sinh ^{2}\left(\sqrt{c}_{1} \xi_{1} / 2\right)\right) /\left(\left(\sqrt{c}_{1}-\sqrt{c}_{2}\right) \cosh \left(\left(\sqrt{c}_{1} \xi_{1}+\sqrt{c}_{2} \xi_{2}\right) / 2\right)\right. \\
\left.+\left(\sqrt{c}_{1}+\sqrt{c}_{2}\right) \cosh \left(\left(\sqrt{c}_{1} \xi_{1}-\sqrt{c}_{2} \xi_{2}\right) / 2\right)\right)^{2}
\end{array}
\label{canhao}
\end{equation}

```python
def double_trouble(t, b, c, c1, c2, n):
    e1 = b - c1*t
    e2 = c - c2*t
    return n*(c1 - c2)*(c1 * pow(np.cosh(np.sqrt(c2)*e2/2), 2) + c2 * pow(np.sinh(np.sqrt(c1)*e1/2), 2))/pow((np.sqrt(c1) - np.sqrt(c2)) * np.cosh((np.sqrt(c1)*e1 + np.sqrt(c2)*e2)/2) + (np.sqrt(c1) + np.sqrt(c2)) * np.cosh((np.sqrt(c1)*e1 - np.sqrt(c2)*e2)/2), 2)
```

```python
x = np.linspace(0, 25, 1000)

plt.axis('off') 
plt.plot(x, double_trouble(x, 7.4, 8, 2, 1, 1), label='fit function')
```
