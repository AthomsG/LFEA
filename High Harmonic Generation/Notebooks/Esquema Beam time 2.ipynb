```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageOps
from matplotlib import rcParams

from mpl_toolkits.mplot3d import Axes3D
#from PIL_Functions import img_to_matrix, matrix_to_img, soma_linhas
```

```python
rcParams['font.size'] = 17
rcParams['axes.labelsize'] = 18
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Computer Modern Roman']
rcParams['text.usetex'] = True
```

```python
##FUNÇÕE AUXILIARES

def img_to_matrix(img):
    width, height = img.size
    pixels = img.load() #returns dictionary where each entry has rgb values

    matrix = list()
    row  = list()
    for y in range(height):
        row = []
        for x in range(width):
            row += [pixels[x, y]]
        matrix.append(row)
    return matrix

def soma_linhas(pixels, orientation='x'):
    height = len(pixels)
    width  = len(pixels[0])

    if orientation=='x':
        soma = list()
        for i in range(width):
            soma += [0]
        for i in range(height):
            for j in range(width):
                soma[j] += pixels[i][j]
        return soma
    else:
        soma = list()
        for i in range(height):
            soma += [0]
        for i in range(width):
            for j in range(height):
                soma[j] += pixels[j][i]
        return soma

def matrix_to_img(matrix):
    height = len(matrix)
    width  = len(matrix[0])

    new_img = Image.new('L',(width, height))

    putdata = list()
    for i in range(height):
        for j in range(width):
            putdata += [matrix[i][j]]

    new_img.putdata(putdata)
    return new_img

def get_max(matrix):
    maximo = 0
    for linha in matrix:
        mx = max(linha)
        if maximo < mx:
            maximo = mx
    return maximo
```

```python
path  = '/Users/thomasgaehtgens/Desktop/LFEA/High Harmonic Generation/Dados/1a sessao/ensaios atenuacao/hhg_Ar12.8mbar_I177_100hz_50deg_100ms.tif'
img   = Image.open(path)
height, width = img.size
scale = 10
```

```python
#Diminuir resolução da imagem, para facilitar processamento
img_pix = img.resize((int(height/scale), int(width/scale)))#, resample=Image.BILINEAR)
# Scale back up using NEAREST to original size
#img     = img_pix.resize(img.size,Image.NEAREST)

matrix = img_to_matrix(img_pix)

x = list()
y = list()
z = list()

for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        x.append(i)
        y.append(j)
        z.append(matrix[i][j])

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')

ax.scatter(np.array(x)*1024/100, np.array(y)*1024/100, z, c=z, s=3)

# Make planes transparent
ax.xaxis.pane.fill = False # Left plane
ax.yaxis.pane.fill = False # Right plane
ax.zaxis.pane.fill = False # Right plane



#VAMOS POR A IMAGEM NO PLOT
X, Y = np.meshgrid(np.linspace(0, len(matrix)-1, len(matrix)), np.linspace(0, len(matrix)-1, len(matrix)))
ax.contourf(Y*10.24, X*10.24, matrix, 100, zdir='z', offset=get_max(matrix), alpha=0.2)

plt.show()
```
